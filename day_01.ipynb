### Создание датасета посещения вет лечебницы

```python
import random
import datetime

vet_visits = {
    'Name': [random.choice(data['Name']) for _ in range(75)],
    'Breed': [random.choice(data['Breed']) for _ in range(75)],
    'Color': [random.choice(data['Color']) for _ in range(75)],
    'Height (cm)': [random.randint(20, 70) for _ in range(75)],
    'Weight (kg)': [random.randint(3, 40) for _ in range(75)],
    'Date of Birth': [f"{random.randint(2015, 2022)}-{random.randint(1, 12):02d}-{random.randint(1, 28):02d}" for _ in range(75)],
    'Visit Date': [datetime.date(2023, random.randint(1, 12), random.randint(1, 28)) for _ in range(75)]
}


vet_visits = pd.DataFrame(vet_visits)

vet_visits = vet_visits[["Visit Date", "Name", "Breed", "Weight (kg)"]]

print(vet_visits)
```

```python
vet_visits.info()
```

```python
unique_dogs = vet_visits.drop_duplicates(subset=['Name', 'Breed'])
```
